```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def plot_altitude_reference():
    # Draw horizontal line segment
    goal_x1, goal_y1, goal_x2, goal_y2 = 0.2, 0, -0.2, 0

    # Clear the current figure
    plt.cla()
    # for stopping simulation with the esc key.
    plt.gcf().canvas.mpl_connect(
        'key_release_event',
        lambda event: [exit(0) if event.key == 'escape' else None])
    # plot_arrow(state.lander_x, state.lander_y, state.lander_theta)
    # plt.plot(cx, cy, "-r", label="course")

    plt.plot([goal_x1, goal_x2], [0, 0], "-g", label="goal", linewidth=5)
    
    # Goal boundary
    plt.axvline(x=goal_x1, color="g", linestyle="--")
    plt.axvline(x=goal_x2, color="g", linestyle="--")

    # # Game boundary
    # plt.axvline(x=-1, color="r", linestyle="--")
    # plt.axvline(x=1, color="r", linestyle="--")

    # Landing Surface
    plt.axhline(y=0, color="k", linestyle="--")

    x = np.linspace(-10, 10, 1000)
    y = np.abs(x)
    plt.plot(x, y, "-b", label="altitude reference")
    
    # Draw arrow

    arrow_x = np.array([-0.8, -0.6, -0.4, -0.2, 0, +0.8, +0.6, +0.4, +0.2])
    label="Lander Goes Up"
    for i in range(len(arrow_x)):
        # arrow_y = np.array([0.5, 0.5, 0.5, 0.5])
        plt.arrow(arrow_x[i], 0, 0, np.abs(arrow_x[i])-0.1, width=0.01, color='red', length_includes_head=True, head_width=0.05, head_length=0.05)
        plt.arrow(arrow_x[i], 1.5, 0, -1.5+(np.abs(arrow_x[i]) + 0.1), width=0.01, color='red', length_includes_head=True, head_width=0.05, head_length=0.05)

    # plt.grid(True)
    # Set y lim
    plt.ylim(-0.5, 1.5)
    plt.xlim(-1.25, 1.25)
    plt.xlabel("Horizontal Position")
    plt.ylabel("Altitude")
    plt.title("Altitude Reference")
    # plt.legend(loc='best', bbox_to_anchor=(1, 0.5, 0.1, 0.5))
    plt.legend(loc='lower left')
    plt.show()

plot_altitude_reference()
```
